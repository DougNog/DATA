```python
#* Para rodar a aplicao
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#* Configurao visual
sns.set_theme(style="darkgrid")

#* Carregar base de dados
df = pd.read_csv('SampleSuperstore.csv')

#* Mostrar as primeiras 5 linhas
display(df.head())

#* Verificando tipos de dados e nulos
df.info()

#* Distribuio de Vendas
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Sales', bins=40, kde=True, color='skyblue')
plt.title('Distribuio de Vendas')
plt.xlabel('Vendas')
plt.ylabel('Contagem')
plt.show()

#* Boxplot de Lucro por Categoria
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Profit', palette='Set2')
plt.title('Distribuio de Lucro por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Lucro')
plt.show()

#* KDE de Lucro
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Profit', fill=True, color='orange')
plt.title('Densidade do Lucro')
plt.xlabel('Lucro')
plt.show()

#* Histograma de Quantidade por Segmento
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Quantity', hue='Segment', multiple='stack', bins=20, palette='pastel')
plt.title('Distribuio de Quantidade por Segmento')
plt.xlabel('Quantidade')
plt.show()

#* Boxplot de Vendas por Regio
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Region', y='Sales', palette='Set3')
plt.title('Vendas por Regio')
plt.xlabel('Regio')
plt.ylabel('Vendas')
plt.xticks(rotation=45)
plt.show()

#* Violin Plot para Sales por Categoria e Segmento
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Category', y='Sales', hue='Segment', split=True, palette='muted')
plt.title('Distribuio de Vendas por Categoria e Segmento')
plt.xlabel('Categoria')
plt.ylabel('Vendas')
plt.show()
```
